# Univariate analysis

#### To analyse the various data sets that are given in the data sets we can start with a univariate analysis to calculate the different measures of central tendancies and dispersion


```python
#Defining a function to ease the process
def diff_plots(data):
    fig, axes =plt.subplots(1,3,figsize=(20, 5))
    Name=data.name.upper()
    fig.suptitle("SPREAD OF DATA FOR "+ Name  , fontsize=15, fontweight='bold')
    sns.distplot(data,kde=True,color='Blue',ax=axes[0])
    axes[0].axvline(data.mean(), color='y', linestyle='--',linewidth=2)
    axes[0].axvline(data.median(), color='r', linestyle='solid', linewidth=2)
    axes[0].axvline(data.mode()[0],color='g',linestyle='dashed',linewidth=2)
    axes[0].legend({'Mean':data.mean(),'Median':data.median(),'Mode':data.mode()})
    sns.boxplot(x=data,showmeans=True, orient='h',color="purple",ax=axes[1])
    #just exploring violin plot
    sns.violinplot(data,ax=axes[2],showmeans=True)
```

```python
diff_plots(data.Education);
```

```python
diff_plots(data.Usage);
```

```python
diff_plots(data.Age);
```

```python
diff_plots(data.Income);
```

```python
diff_plots(data.Fitness);
```

```python
diff_plots(data.Miles);
```

#### We analysed the quantitative values, lets analyse the categorical values:

```python
plt.figure(figsize=(15,10))
explode = (0.1, 0.1, 0.1)
data['Product'].value_counts().plot.pie(autopct='%1.1f%%',figsize=(5,5),explode=explode)
plt.title("Product Sales",size=15,style="italic")
plt.show()
```

```python
plt.figure(figsize=(10,6))
ax =sns.countplot(data=data,y='Product');
total = len(data['Product'])
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_width()/total)
    x = p.get_x() + p.get_width() + 0.02
    y = p.get_y() + p.get_height()/2
    ax.annotate(percentage, (x,y))
plt.title("Product Sales",size=15,style="italic")
plt.show()
```

#### Inference
* The most common product that is sold is TM195 with 44.4% of total sales
